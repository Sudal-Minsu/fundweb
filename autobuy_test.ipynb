```python
import requests
import json
import random
import time
import matplotlib.pyplot as plt

# üìå Matplotlib Ïù∏ÌÑ∞ÎûôÌã∞Î∏å Î™®Îìú ÌôúÏÑ±Ìôî (Ïã§ÏãúÍ∞Ñ ÏóÖÎç∞Ïù¥Ìä∏)
plt.ion()

# üîπ API Key ÏÑ§Ï†ï (Î™®ÏùòÍ≥ÑÏ¢å)
app_key = keyring.get_password('mock_app_key', 'ÏµúÏßÑÌòÅ')
app_secret = keyring.get_password('mock_app_secret', 'ÏµúÏßÑÌòÅ')

# üîπ Î™®ÏùòÌà¨Ïûê API Í∏∞Î≥∏ URL
url_base = "https://openapivts.koreainvestment.com:29443"

# üîπ Ï†ëÍ∑ºÌÜ†ÌÅ∞ Î∞úÍ∏â
headers = {"content-type": "application/json"}
path = "oauth2/tokenP"
body = {
    "grant_type": "client_credentials",
    "appkey": app_key,
    "appsecret": app_secret
}
url = f"{url_base}/{path}"

res = requests.post(url, headers=headers, data=json.dumps(body))
access_token = res.json().get('access_token', '')
print(f"üîë Î∞úÍ∏âÎêú Ïï°ÏÑ∏Ïä§ ÌÜ†ÌÅ∞: {access_token}")

# üîπ ÎûúÎç§ Ï¢ÖÎ™© Î¶¨Ïä§Ìä∏ (ÌïúÍµ≠ Ï£ºÏãù)
stock_list = ["005930", "000660", "035420", "068270", "028260"]  # ÏÇºÏÑ±Ï†ÑÏûê, SKÌïòÏù¥ÎãâÏä§, NAVER, ÏÖÄÌä∏Î¶¨Ïò®, ÏÇºÏÑ±Î¨ºÏÇ∞

# üîπ Í±∞Îûò ÎÇ¥Ïó≠ Ï†ÄÏû•
trade_history = {}
profit_log = []
profit_rate_log = []

# üîπ Ìï¥ÏãúÌÇ§ ÏÉùÏÑ± Ìï®Ïàò
def get_hashkey(data):
    path = "uapi/hashkey"
    url = f"{url_base}/{path}"
    
    headers = {
        "Content-Type": "application/json",
        "appKey": app_key,
        "appSecret": app_secret
    }
    
    res = requests.post(url, headers=headers, data=json.dumps(data))
    return res.json().get("HASH", "")

# üîπ ÌòÑÏû¨ Ï£ºÍ∞Ä Ï°∞Ìöå Ìï®Ïàò
def get_current_price(stock_code):
    path = "uapi/domestic-stock/v1/quotations/inquire-price"
    url = f"{url_base}/{path}"

    headers = {
        "Content-Type": "application/json",
        "authorization": f"Bearer {access_token}",
        "appKey": app_key,
        "appSecret": app_secret,
        "tr_id": "FHKST01010100"
    }

    params = {"fid_cond_mrkt_div_code": "J", "fid_input_iscd": stock_code}
    res = requests.get(url, headers=headers, params=params)
    
    # üîπ API ÏùëÎãµ Ïò§Î•ò Ï≤òÎ¶¨
    if 'output' not in res.json():
        print(f"‚ö†Ô∏è API Ïò§Î•ò: {res.json()}")
        return None
    
    return int(res.json()['output']['stck_prpr'])  # ÌòÑÏû¨ Ï£ºÍ∞Ä Î∞òÌôò

# üîπ ÎûúÎç§ Îß§Îß§ Ìï®Ïàò
def random_trade():
    stock_code = random.choice(stock_list)  # üîπ ÎûúÎç§ Ï¢ÖÎ™© ÏÑ†ÌÉù
    order_type = random.choice(["BUY", "SELL"])  # üîπ ÎûúÎç§ÏúºÎ°ú Îß§Ïàò/Îß§ÎèÑ ÏÑ†ÌÉù
    quantity = random.randint(1, 10)  # üîπ ÎûúÎç§ Ï£ºÎ¨∏ ÏàòÎüâ (1~10Ï£º)
    
    print(f"üõí Ï¢ÖÎ™©: {stock_code}, Ï£ºÎ¨∏: {order_type}, ÏàòÎüâ: {quantity}Ï£º")

    path = "/uapi/domestic-stock/v1/trading/order-cash"
    url = f"{url_base}/{path}"
    
    data = {
        "CANO": "50124996",  # Í≥ÑÏ¢åÎ≤àÌò∏ Ïïû 8ÏûêÎ¶¨
        "ACNT_PRDT_CD": "01",  # Í≥ÑÏ¢åÎ≤àÌò∏ Îí§ 2ÏûêÎ¶¨
        "PDNO": stock_code,  # Ï¢ÖÎ™©ÏΩîÎìú
        "ORD_DVSN": "01",  # Ï£ºÎ¨∏ Î∞©Î≤ï (ÏãúÏû•Í∞Ä Ï£ºÎ¨∏)
        "ORD_QTY": str(quantity),  # Ï£ºÎ¨∏ ÏàòÎüâ
        "ORD_UNPR": "0",  # ÏãúÏû•Í∞Ä Ï£ºÎ¨∏ (0)
    }

    headers = {
        "Content-Type": "application/json",
        "authorization": f"Bearer {access_token}",
        "appKey": app_key,
        "appSecret": app_secret,
        "tr_id": "VTTC0802U" if order_type == "BUY" else "VTTC0801U",  # üîπ Îß§Ïàò/Îß§ÎèÑ tr_id ÏÑ§Ï†ï
        "custtype": "P",
        "hashkey": get_hashkey(data)
    }

    res = requests.post(url, headers=headers, data=json.dumps(data))

    if res.json().get('rt_cd') == '0':  # Ï†ïÏÉÅ Ï£ºÎ¨∏ ÏùëÎãµ ÏΩîÎìú
        print(f"‚úÖ {order_type} Ï£ºÎ¨∏ ÏÑ±Í≥µ: {res.json()}")
        
        # üîπ Îß§Ïàò Ïãú, Íµ¨Îß§ Í∞ÄÍ≤© Ï†ÄÏû•
        if order_type == "BUY":
            trade_history[stock_code] = get_current_price(stock_code)

        # üîπ Îß§ÎèÑ Ïãú, ÏàòÏùµ Í≥ÑÏÇ∞
        elif order_type == "SELL" and stock_code in trade_history:
            buy_price = trade_history[stock_code]
            sell_price = get_current_price(stock_code)

            if buy_price and sell_price:
                profit = (sell_price - buy_price) * quantity
                profit_rate = ((sell_price - buy_price) / buy_price) * 100

                profit_log.append(profit)
                profit_rate_log.append(profit_rate)

                print(f"üí∞ ÏàòÏùµ: {profit}Ïõê, ÏàòÏùµÎ•†: {profit_rate:.2f}%")
    
    else:
        print(f"‚ö†Ô∏è Ï£ºÎ¨∏ Ïã§Ìå®: {res.json()}")

# üîπ Ïã§ÏãúÍ∞Ñ ÏàòÏùµÎ•† Í∑∏ÎûòÌîÑ Ìï®Ïàò
def plot_profit():
    plt.clf()
    plt.figure(figsize=(10, 5))

    # üîπ Ï¥ù ÏàòÏùµ (Ïõê) Í∑∏ÎûòÌîÑ
    plt.subplot(2, 1, 1)
    plt.plot(profit_log, marker='o', linestyle='-', color='b', label="Ï¥ù ÏàòÏùµ (Ïõê)")
    plt.xlabel("Í±∞Îûò ÌöüÏàò")
    plt.ylabel("Ï¥ù ÏàòÏùµ (Ïõê)")
    plt.legend()
    plt.grid()

    # üîπ ÏàòÏùµÎ•† (%) Í∑∏ÎûòÌîÑ
    plt.subplot(2, 1, 2)
    plt.plot(profit_rate_log, marker='o', linestyle='-', color='r', label="ÏàòÏùµÎ•† (%)")
    plt.xlabel("Í±∞Îûò ÌöüÏàò")
    plt.ylabel("ÏàòÏùµÎ•† (%)")
    plt.legend()
    plt.grid()

    plt.tight_layout()
    plt.pause(0.1)

# üîπ ÎûúÎç§ Îß§Îß§ Ïã§Ìñâ (Ïã§ÏãúÍ∞Ñ Í∑∏ÎûòÌîÑ ÏóÖÎç∞Ïù¥Ìä∏)
start_time = time.time()
while True:
    try:
        random_trade()  # üîπ ÎûúÎç§ÏúºÎ°ú Îß§Îß§ Ïã§Ìñâ

        # üîπ 1Î∂ÑÎßàÎã§ Í∑∏ÎûòÌîÑ ÏóÖÎç∞Ïù¥Ìä∏
        if (time.time() - start_time) > 60:
            plot_profit()
            start_time = time.time()

        time.sleep(random.randint(5, 10))  # üîπ 5~10Ï¥à ÎåÄÍ∏∞ ÌõÑ Îã§Ïùå Í±∞Îûò Ïã§Ìñâ

    except Exception as e:
        print(f"‚ùå Ïò§Î•ò Î∞úÏÉù: {e}")
        time.sleep(5)  # 5Ï¥à ÌõÑ Ïû¨ÏãúÎèÑ
```
